```python
import turtle
import random
import time

# ---------------- Screen setup ----------------
screen = turtle.Screen()
screen.setup(width=1.0, height=1.0)
screen.title("Merry Christmas - Python Turtle")
screen.bgcolor("midnight blue")
screen.tracer(0)

# ---------------- Tree configuration ----------------
tree_length = 120
turtle.speed("fastest")
turtle.hideturtle()
turtle.left(90)
turtle.penup()
turtle.backward(tree_length * 2)
turtle.pendown()
turtle.color("dark green")

# Draw Christmas tree using recursion
def draw_tree(depth, length):
    if depth == 0:
        return
    turtle.forward(length)
    turtle.left(30)
    draw_tree(depth - 1, length * 0.75)
    turtle.right(60)
    draw_tree(depth - 1, length * 0.75)
    turtle.left(30)
    turtle.backward(length)

draw_tree(10, tree_length)

# ---------------- Greeting text ----------------
pen = turtle.Turtle()
pen.hideturtle()
pen.penup()
pen.color("gold")
pen.goto(-250, -250)
pen.write(
    "Merry Christmas!",
    font=("Arial", 36, "bold")
)

# ---------------- Snowflake setup ----------------
snowflakes = []

def create_snow():
    for _ in range(40):
        snow = turtle.Turtle()
        snow.hideturtle()
        snow.penup()
        snow.speed(0)
        snow.color("white")
        snow.shape("circle")
        snow.shapesize(0.3)
        snow.goto(
            random.randint(-500, 500),
            random.randint(0, 400)
        )
        snow.showturtle()
        snowflakes.append(snow)

def snowfall():
    for snow in snowflakes:
        y = snow.ycor()
        y -= random.randint(1, 5)
        snow.sety(y)

        # Reset snowflake to top when it reaches bottom
        if y < -350:
            snow.goto(
                random.randint(-500, 500),
                random.randint(300, 500)
            )

# Create snowflakes
create_snow()

# ---------------- Animate snowfall ----------------
while True:
    snowfall()
    screen.update()
    time.sleep(0.05)
```
